```python
def generate_random_numbers(shape, distribution="gaussian", seed=None):
    """
    The function should support both uniform and Gaussian distributions.

    Args:
        shape (int or tuple of ints):
            Desired shape of the output.
            - If int, return a list of length `shape`
            - If tuple, return nested lists with that shape
        distribution (str):
            Type of distribution to generate.
            Supported values: "uniform", "gaussian".
        seed (int, optional):
            Seed value to ensure reproducibility. If None, the seed
            may be derived from the current time.

    Returns:
        list:
            Random numbers shaped according to `shape`.
            Values should lie in:
            - [0, 1) for uniform distribution
            - approximately zero mean and unit variance for Gaussian
    """
    # YOUR CODE HERE
    if (seed == None):
        seed = int(time.time())

    a = 1389796
    c = 0
    m = (2 ** 32) - 5
    num = seed
    if distribution == "uniform":
        result = []
        if (type(shape) is int):
            for i in range(shape):
                num = (a * num + c) % m        
                result.append(num/m)
        elif (type(shape) is tuple):
            totalNums = 1
            for i in shape:
                totalNums *= i
            for i in range(totalNums):
                num = (a * num + c) % m        
                result.append(num/m)

            result = np.reshape(result, shape)
            result = result.tolist()
    elif distribution == "gaussian":
        result = []

        seed1 = seed
        seed2 = seed1 + 32443

        num1 = seed1
        num2 = seed2

        if (type(shape) is int):
            for i in range(shape):
                num1 = (a * num1 + c) % m
                num2 = (a * num2 + c) % m
                gausVar = ( np.sqrt(-2 * np.log(num1 / m) )) * (np.cos(2 * np.pi * num2 / m))
                result.append(gausVar)
        elif (type(shape) is tuple):
            totalNums = 1
            for i in shape:
                totalNums *= i
            for i in range(totalNums):
                num1 = (a * num1 + c) % m   
                num2 = (a * num2 + c) % m  
                gausVar = ( np.sqrt(-2 * np.log(num1 / m) )) * (np.cos(2 * np.pi * num2 / m))
                result.append(gausVar)

            result = np.reshape(result, shape)
            result = result.tolist()

    return result
```

```python
def min_max_normalize(image):
    """
    Apply min-max normalization to an image.

    Args:
        image (np.ndarray): Input image.

    Returns:
        np.ndarray: Normalized image with values in [0, 1].
    """
    # YOUR CODE HERE

    
    ##################### Div by zero
    mini = image.min()
    maxi = image.max()
    image = (image - mini) / (maxi - mini)
    return image
```

```python
def standardize(image, mean=0, std=1):
    """
    Standardize image pixel values using a specified mean and standard deviation.

    Args:
        image (np.ndarray): Input image.
        mean (float): Desired mean of the standardized output.
        std (float): Desired standard deviation of the standardized output.

    Returns:
        np.ndarray: Standardized image array.
    """
    # YOUR CODE HERE



    ######### CHeck size

    
    standardised = (image - image.mean()) / image.std()
    standardised = (standardised * std) + mean

    return standardised
```

```python
def crop_image(image, top, left, height, width):
    """
    Crop a rectangular region from an image using NumPy slicing.

    This function should return a view (not a copy) of the original image
    whenever possible, demonstrating how NumPy handles memory efficiently.

    Args:
        image (np.ndarray):
            Input image.
        top, left (int):
            Top-left corner of the crop.
        height, width (int):
            Size of the cropped region.

    Returns:
        np.ndarray:
            Cropped image.

    """
    # YOUR CODE HERE

    return image[top: top + height, left: left + width]
```

```python
def resize_with_padding(image, target_size=(224, 224)):
    """
    Resize an image to a fixed target size while preserving aspect ratio.

    The image should be resized such that its aspect ratio is preserved.
    Any remaining area should be padded with zeros (black pixels) to
    achieve the target size.

    This function must work for both:
    - Color images of shape (H, W, 3)
    - Grayscale images of shape (H, W)

    Args:
        image (np.ndarray): Input image as a NumPy array.
        target_size (tuple): Desired output size as (height, width).
            Default is (224, 224).

    Returns:
        np.ndarray: Resized image of shape:
            - (224, 224, 3) for color images
            - (224, 224) for grayscale images
    """
    # YOUR CODE HERE

    scalingFactor = min(target_size[0]/image.shape[0], target_size[1]/image.shape[1])
    newHeight = int(image.shape[0] * scalingFactor)
    newWidth = int(image.shape[1] * scalingFactor)

    resizedImage = cv2.resize(image, (newWidth, newHeight))

    resizedImage = cv2.copyMakeBorder(resizedImage, 0, target_size[0] - newHeight, 0, target_size[1] - newWidth, cv2.BORDER_CONSTANT, value=[0,0,0])

    return resizedImage
```

```python
def extract_rgb_channels(image):
    """Extract R, G, B channels."""
    # YOUR CODE HERE
    r, g, b = cv2.split(image)
    return (r, g, b)
```

```python
def rgb_to_grayscale(image):
    """
    Convert RGB image to grayscale and normalize pixel values to [0, 1].
    """
    # YOUR CODE HERE
    grayscaleImage = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

    fig, ax = plt.subplots()
    ax.set_title("Gray")
    ax.hist(grayscaleImage.flatten(), bins=256, edgecolor="black", color="gray")

    normalisedImage = grayscaleImage / 255
    return normalisedImage
```

```python
def add_gaussian_noise(image, mean=0.0, std=0.05):
    """Add Gaussian noise using NumPy."""
    # YOUR CODE HERE

    normalisedImg = min_max_normalize(image)
    imgWithNoise = normalisedImg + np.random.normal(mean, std)
    finalImage = (imgWithNoise * 255).astype(np.uint8)
    return finalImage
```

```python
def impute_missing_values(df):
    """
    Handle missing values in the dataset.

    - Numerical features should be imputed using a statistical strategy
      (e.g., mean or median).
    - Categorical features should be imputed using an appropriate method
      (e.g., mode).

    Args:
        df (pd.DataFrame): Input dataset with missing values.

    Returns:
        pd.DataFrame: Dataset after missing value imputation.
    """
    # YOUR CODE HERE
    numericalColumns = df.select_dtypes(include=['number']).columns
    for i in numericalColumns:
        if df[i].isnull().values.any():
            columnMedian = df[i].median()
            df[i] = df[i].fillna(columnMedian)
            
    categoricalColumns = df.select_dtypes(include=['object', 'str', 'category']).columns
    for i in categoricalColumns:
        if df[i].isnull().values.any():
            columnMode = df[i].mode()[0]
            df[i] = df[i].fillna(columnMode)
    print("Replaced missing numerical values with their mean, and categorical values with their mode")

    return df
```

I used median for imputation of numerical columns. The mean is highly sensitive to outliers; a single extremely high or low value can skew the average significantly, making it an inaccurate representation of the "center." The median (the middle value) is resistant to these extremes.


For categorical features, i used mode, since we cannot compute mean and median for non-numeric data


```python
def remove_duplicate_rows(df):
    """
    Detect and remove duplicate rows from the dataset.

    The function should also report the number of duplicate rows removed.

    Args:
        df (pd.DataFrame): Input dataset.

    Returns:
        pd.DataFrame: Dataset after removing duplicates.
    """
    # YOUR CODE HERE
    print("Number of duplicate rows:", df.duplicated().sum())
    df = df.drop_duplicates()
    return df
```

```python
def detect_and_handle_outliers(df, numerical_columns):
    """
    Detect outliers in numerical features using IQR or Z-score method.

    - Visualize outliers using boxplots.
    - Decide whether to remove or cap outliers.

    Args:
        df (pd.DataFrame): Input dataset.
        numerical_columns (list): List of numerical feature names.

    Returns:
        pd.DataFrame: Dataset after outlier handling.
    """
    # YOUR CODE HERE
    fig, ax = plt.subplots(len(numerical_columns), 1, figsize=(5, len(numerical_columns) * 5))

    k = 0
    for i in numerical_columns:
        data = df[i]
        sns.boxplot(x=data, ax=ax[k], color='blue')
        k += 1

    plt.show()

    for i in numerical_columns:
        data = df[i]
        q1 = np.quantile(data, 0.25)
        q3 = np.quantile(data, 0.75)
        iqr = q3 - q1
        lowerBound = q1 - 1.5 * iqr
        upperBound = q3 + 1.5 * iqr
        df[i] = df[i].clip(lower=lowerBound, upper=upperBound)

    return df  
```

I preffered to clip, instead of deleting the row entirely. Removing rows with outliers reduces your dataset size, which can be costly for small datasets. A row might contain an outlier in one feature but perfectly valid and valuable signals in all other features. Deleting the entire row destroys that valuable information.

```python
def preprocess_features(df, target_column):
    """
    Encode categorical features and normalize numerical features.

    - Categorical variables should be encoded using appropriate techniques.
    - Numerical features should be scaled using Min-Max or Standard Scaling.

    Args:
        df (pd.DataFrame): Input dataset.
        target_column (str): Name of the target column.

    Returns:
        X (np.ndarray): Preprocessed feature matrix.
        y (np.ndarray): Target vector.
    """
    # YOUR CODE HERE
    X = df.drop(columns=[target_column])
    y = df[target_column].values.tolist()

    numericalColumns = X.select_dtypes(include=['number']).columns
    categoricalColumns = X.select_dtypes(include=['object', 'str', 'category']).columns

    scaler = StandardScaler()
    X_numeric_scaled = scaler.fit_transform(X[numericalColumns])

    ohEncoder = OneHotEncoder(sparse_output=False, drop='first')
    X_categorical_oh = ohEncoder.fit_transform(X[categoricalColumns])

    X_final = np.hstack((X_numeric_scaled, X_categorical_oh))

    return X_final, y
```

```python
def stratified_split_with_smote(X, y, test_size=0.2, random_state=42):
    """
    Perform a stratified train-test split and apply SMOTE on the training set.

    SMOTE must be applied ONLY on the training data.

    Args:
        X (np.ndarray): Feature matrix.
        y (np.ndarray): Target labels.
        test_size (float): Fraction of data used for testing.
        random_state (int): Random seed.

    Returns:
        X_train_resampled, X_test, y_train_resampled, y_test
    """
    # YOUR CODE HERE
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state, stratify=y)

    smote = SMOTE(sampling_strategy="minority")
    X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

    return X_train_resampled, X_test, y_train_resampled, y_test
```

```python
def split_regression_data(X, y, test_size=0.2, random_state=42):
    """
    Split regression data into training and testing sets.

    Args:
        X (np.ndarray): Feature matrix.
        y (np.ndarray): Target values.
        test_size (float): Fraction used for testing.
        random_state (int): Random seed.

    Returns:
        X_train, X_test, y_train, y_test
    """
    # YOUR CODE HERE
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
    return X_train, X_test, y_train, y_test
```

```python
def train_linear_regression_model(X_train, y_train):
    """
    Train a linear regression model.

    The model may be trained using:
    - Closed-form solution, or
    - A standard machine learning library.

    Args:
        X_train (np.ndarray): Training features.
        y_train (np.ndarray): Training targets.

    Returns:
        model: Trained linear regression model.
    """
    # YOUR CODE HERE
    from sklearn import linear_model
    reg = linear_model.LinearRegression()
    reg.fit(X_train, y_train)
    return reg
```

```python
def evaluate_linear_regression(model, X_test, y_test):
    """
    Evaluate the trained linear regression model.

    This function should compute:
    - Mean Squared Error (MSE)
    - Root Mean Squared Error (RMSE)
    - R^2 score

    Args:
        model: Trained regression model.
        X_test (np.ndarray): Test features.
        y_test (np.ndarray): Test targets.

    Returns:
        None
    """
    # YOUR CODE HERE
    y_predicted = model.predict(X_test)
    rss = ((y_test - y_predicted) ** 2).sum()
    mse = rss/ len(y_test)
    rmse = np.sqrt(mse)
    tss = ((y_test - np.mean(y_test)) ** 2).sum()
    r2Score = 1 - (rss/tss)

    print("Mean Squared Error:", mse)
    print("Root Mean Squared Error:", rmse)
    print("R^2 Score:", r2Score)
```

```python
def interpret_linear_model_coefficients(model, feature_names):
    """
    Interpret coefficients of the linear regression model.

    Args:
        model: Trained linear regression model.
        feature_names (list): Names of input features.

    Returns:
        None
    """
    # YOUR CODE HERE
    coef = model.coef_
    featureWithCoef = zip(feature_names, coef)
    for i,j in featureWithCoef:
        print(f"Feature {i}: Coefficient {j}")
    
```
